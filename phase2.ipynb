```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Spending_USD'], kde=True)
plt.show()

sns.boxplot(x=df['Spending_USD'])
plt.show()
```

```python
df["Life_Expectancy"].describe()
```

```python
sns.histplot(df["Life_Expectancy"], kde=True)
plt.show()

sns.boxplot(x=df["Life_Expectancy"])
plt.show()
```

```python
sns.lineplot(data=df, x="Year", y="Life_Expectancy", hue="Country")
plt.show()
```

```python
# Filtrage
comparison_df = df[df['Country'].isin(['Germany', 'Great Britain'])] # Adapte selon les noms exacts

plt.figure(figsize=(10, 5))
sns.lineplot(data=comparison_df, x='Year', y='Spending_USD', hue='Country')
plt.title('Comparaison des dépenses de santé : Allemagne vs UK')
plt.show()
```
