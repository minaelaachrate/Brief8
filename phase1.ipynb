1. Statistiques descriptives
Moyenne:La moyenne est la somme de toutes les valeurs divisée par le nombre d’observations.
Médiane:La médiane est la valeur centrale après avoir trié les données.
Variance et Écart-type: L’écart-type mesure la dispersion des données autour de la moyenne.
Histogramme:Un histogramme permet de visualiser la distribution des données.
Boxplot:Un boxplot permet de visualiser :
    la médiane
    les quartiles
    les valeurs aberrantes


2. Population vs Échantillon
Population : ensemble complet des données (ex : tous les pays du monde)
Échantillon : une partie de la population (ex : 50 pays étudiés)
Pourquoi utiliser un échantillon ?
        .Gain de temps
        .Réduction des coûts
        .Analyse plus simple


3. Probabilités et distributions
    _Loi de Bernoulli:
    Modélise une expérience avec deux résultats : succès ou échec.
    Ex:
    Un patient survit ou ne survit pas.
    _Loi Binomiale
    Compte le nombre de succès dans plusieurs essais.
    Ex:
    Sur 100 patients, combien survivent ?
    _Loi Normale
    Distribution en forme de cloche.
    La majorité des valeurs sont proches de la moyenne.
    _Loi de Student
    Utilisée pour les petits échantillons.
    Ex:
    Analyse de seulement 10 pays.

4-Principe du TCL
_Le Théorème Central Limite dit :
Même si la distribution des données initiales n’est pas normale, la distribution des moyennes d’échantillons suffisamment grands devient approximativement normale.
_Illustration intuitive
.Données initiales (non normales)
    Par exemple, les revenus ou les dépenses médicales sont souvent asymétriques.
.Moyennes d’échantillons
    .On prend des échantillons de 50 personnes
    .On calcule la moyenne de chaque échantillon
    .On fait un histogramme de ces moyennes
_Importance du TCL
Permet d’utiliser intervalles de confiance et tests d’hypothèses même si les données ne sont pas normales.

5-Échantillonnage et estimation : définition, biais possibles, notion d’estimation
    1-1_Échantillonnage : définition
    choisir un sous-ensemble d’individus ou d’éléments dans une population pour étudier ses caractéristiques
    1-2_Estimation : définition
    utiliser les données de l’échantillon pour approcher une valeur inconnue de la population.
    1-3_Biais possibles
    Même si l’échantillon est calculé, il peut y avoir des erreurs systématiques
   

6-test d’hypothèse:
    C’est une méthode en statistique pour prendre une décision basée sur des données.
        .Tu veux vérifier si quelque chose est vrai ou faux.
2. H0 et H1 (les hypothèses)
    _H0 (hypothèse nulle)
         C’est la situation normale (pas de changement)
    _H1 (hypothèse alternative)
         C’est ce que tu veux prouver
3. La p-value 
    .Est-ce que le résultat que j’ai est bizarre ou normal ?
            petite p-value → résultat rare → H0 faux
            grande p-value → résultat normal → H0 vrai
4. Interprétation (comment comprendre)
    Tu compares la p-value avec 0.05
        p-value     	décision
        ≤ 0.05	        on rejette H0
        > 0.05	        on garde H0
